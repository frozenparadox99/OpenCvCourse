```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
naadia=cv2.imread('Computer-Vision-with-Python/DATA/Nadia_Murad.jpg',0)
denis=cv2.imread('Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg',0)
solvay=cv2.imread('Computer-Vision-with-Python/DATA/solvay_conference.jpg',0)
```

```python
face_cascade=cv2.CascadeClassifier('Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img=img.copy()
    
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result=detect_face(naadia)
plt.imshow(result,cmap='gray')
```

```python
result=detect_face(denis)
plt.imshow(result,cmap='gray')
```

```python
result=detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    frame=detect_face(frame)
    cv2.imshow('WindowName',frame)
    
    k=cv2.waitKey(1)
    if k & 0xFF==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
